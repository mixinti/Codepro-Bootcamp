```python
base_ejemplo = {"nombre" : "base", 
               "precio" : 40000, 
               "stock" : True}

print(base_ejemplo)
base_ejemplo["nombre"]
print(base_ejemplo["precio"])

#modificar

base_ejemplo["precio"] = 45000
print(base_ejemplo["precio"])

#metodos
print(base_ejemplo.keys()) #devuelve todas las claves
print(base_ejemplo.values()) #devuelve toos los valores
print(base_ejemplo.items()) #devuelve el par

#anhadir
base_ejemplo["tienda"] = "mac"
print(base_ejemplo)

#eliminar el ultimo elemento (clave-valor) del diccionario
base_ejemplo.popitem()
print(base_ejemplo)

#eliminar dato del dicionario x clave
del base_ejemplo["precio"]
print(f"diccionario con la clave eliminada: {base_ejemplo}")

#los diccionarios son iterables
for i in base_ejemplo.values():
    print(i)
```

```python
#Challenge 1

info_persona = {} #creo la lista vacia

informacion = ["nombre", "edad", "sexo", "telefono", "correo", "residencia"] #defino la informacion a preguntar

for info in informacion: #pregunto al usuario x la info
    pregunta = input(f"ingrese su {info}: ")
    info_persona[info] = pregunta

print(info_persona)
```

```python
#Challenge 2

tarjeta={}

while True:
    accion=int(input("que desea realizar,1=crear,2=modificar,3=eliminar "))
  
    if accion==1:
        clave_1=input("ingrese la clave")
        valor_1=input("ingrese el valor para la clave")
        tarjeta[clave_1]=valor_1
        print(tarjeta)
    elif accion==2:
        print("los valores que podes modificar son son")
        print(tarjeta.items())
        clave_1=input("ingrese la clave a cambiar")
        modificacion=input("ingrese el valor nuevo")
        tarjeta[clave_1]=modificacion
        print(tarjeta)
    elif accion==3:
        print("los valores que podes eliminar son")
        print(tarjeta.keys())
        clave_1=input("ingresa el valor a eliminar")
        del tarjeta[clave_1]
        print(tarjeta)
    parar=input("si desea salir ingrese si sino ingrese no")
    if parar=="si":
        break
```

```python
#Challenge 3

agenda = {}

opcion = ""

while opcion != "4":
    print("1. Mostrar contactos")
    print("2. Crear contacto")
    print("3. Modificar contacto")
    print("4. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        print(agenda)

    elif opcion == "2":
        nombre = input("Nombre: ")
        telefono = input("Teléfono: ")
        agenda[nombre] = telefono
        print(f"Contacto guardado: {nombre, telefono}")

    elif opcion == "3":
        nombre = input("Nombre del contacto a modificar: ")
        if nombre in agenda:
            telefono = input("Nuevo teléfono: ")
            agenda[nombre] = telefono
            print(f"Contacto modificado: {nombre, telefono}")
        else:
            print("Contacto no existe")

    elif opcion == "4":
        print("Saliendo...")

    else:
        print("Opción inválida")
```

```python
#Challenge 4

class Animales:
    especie = "Ave"
    def __init__(self, nombre, color): #defino funcion para atributos de instancia
        self.nombre = nombre
        self.color = color
    def presentarse(self): #defino funcion para presentarse
        print(f"Me llamo {self.nombre} y soy un ave de color {self.color}")

ave1 = Animales("Cuervo", "Negro")
ave2 = Animales("Paloma", "Gris")
ave3 = Animales("Flamenco", "Rosado")

ave1.presentarse()
ave2.presentarse()
ave3.presentarse()
```

```python
class Superheroe:
    especie = "superhumano"
#metodo constructor
    def __init__(self, nombre):
        self.nombre = nombre

    def rayos_laser(self):
        print("tiro rayos laser rojos")

    def volar(self, km):
        print(f"estoy volando a {km} km por hora")

    def alzar_objeto(self, objeto, kg):
        print(f"puedo alzar un {objeto} de {kg} kg")
    
#instanciar objetos
superheroe1 = Superheroe("Batman")
superheroe1.volar(120)

superheroe2 = Superheroe("Spiderman")
superheroe2.alzar_objeto("tren", "1000")
```

```python
class AnimalDeGranja:
    tipo = "animal de granja"

    def __init__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido

    def hablar(self):
        print(f"Hablo y digo {self.sonido}")
    
    def presentacion(self):
        print(f"Hola, soy {self.nombre}")

class Gallina(AnimalDeGranja):
    huevos_por_dia = 1

    def poner_huevo(self):
        print(f"Estoy poniendo {self.huevos_por_dia} huevos por dia")

class Vaca(Gallina):
    patas = 4
    
    def mujir(self):
        print("Muu")

gallina1 = Gallina("Hortencia", "marron", "pio pio")
gallina1.hablar()
gallina1.presentacion()

vaca1 = Vaca("Lola", "blanco y negro", "Muu")
vaca1.hablar()
vaca1.mujir()
```

```python
#Challenge 6

class Maquillaje:
    def __init__(self, marca, precio):
        self.marca = marca
        self.precio = precio

class Rimel(Maquillaje):
    calidad = "buena"

class Labial(Maquillaje):
    calidad = "buena"

rimel1 = Rimel("Maybelline", 75000)
labial1 = Labial("Mac", 55000)

print(f"El rimel es de marca {rimel1.marca}, su precio es de {rimel1.precio} y tiene calidad {rimel1.calidad}")
print(f"El labial es de marca {labial1.marca}, su precio es de {labial1.precio} y tiene calidad {labial1.calidad}")
```
